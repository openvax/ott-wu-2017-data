```python
peptide_dict_supp4 = dict(zip(df_supp4["Immunizing peptide ID"], df_supp4["Immunizing peptide sequence"]))
```

```python
len(peptide_dict_supp4)
```

```python
from collections import Counter
supp4_counts = Counter()
for peptide_id in peptide_dict_supp4:
    supp4_counts[peptide_id[0]] += 1
supp4_counts
```

```python
df_counts = pd.read_csv("Supplemental information 3.csv")
```

```python
# do the counts for each patient match our expections?
for (patient, count) in zip(df_counts["Patient"], df_counts["Number of peptides included in vaccine"]):
    patient = str(patient)
    print(patient)
    if patient in supp4_counts:
        print("\t %d %d" % (supp4_counts[patient], int(count)))
        assert supp4_counts[patient] == int(count)
```

```python
with open("immunizing-peptides.csv", "w") as f:
    f.write("Patient,Immunizing peptide ID,Immunizing peptide sequence\n")
    for peptide_id in sorted(peptide_dict_supp4):
        peptide = peptide_dict_supp4[peptide_id]
        f.write("%s,%s,%s\n" % (peptide_id[0], peptide_id, peptide))
```
